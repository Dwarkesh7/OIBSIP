```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Unemployment in India.csv'
data = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("First few rows of the dataset:")
print(data.head())






```

```python

print("\nSummary statistics of the dataset:")
print(data.describe())
```

```python
print("\nData types of each column:")
print(data.dtypes)
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
print("Column names in the dataset:")
print(data.columns)
```

```python
# Handle missing values if any (this example assumes no missing values or simple dropna for simplicity)
data = data.dropna()
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
data.columns = data.columns.str.strip()
```

```python
# Time-series analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Estimated Employed', data=data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Employed')
plt.xticks(rotation=45)
plt.show()
```

```python
# Time-series analysis
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Frequency', data=data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Geographical analysis (assuming there are columns like 'State' and 'Unemployment Rate')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Estimated Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate by State')
plt.xlabel('State')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Additional analyses based on the columns available in the dataset
plt.figure(figsize=(12, 6))
sns.barplot(x='Area', y='Estimated Labour Participation Rate (%)', data=data)
plt.title('Unemployment Rate by Area(Rural/Urban)')
plt.xlabel('Gender')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.show()
```
